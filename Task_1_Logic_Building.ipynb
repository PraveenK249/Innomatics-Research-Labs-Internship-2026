```python
# 1. User Login Check

username = "admin"
password = "1234"

input_username = input("Enter the username: ")   # Example input
input_password = input("Enter the password: ")

if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# 2. Pass / Fail Analyzer

# List of student marks
marks = [45, 78, 90, 33, 60]

# Counters for pass and fail students
pass_count = 0
fail_count = 0

# Loop through each student's marks
for mark in marks:
    # A student passes if marks >= 50
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print final results
print("\nPass Students:", pass_count)
print("Fail Students:", fail_count)
```

```python
# 3. Simple Data Cleaner

# Raw names list with spaces and inconsistent formatting
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

# Clean each name
for name in names:
    # strip() removes extra spaces
    # lower() converts to lowercase
    cleaned_names.append(name.strip().lower())

# Display cleaned data
print("\nCleaned Names:", cleaned_names)
```

```python
# 4. Message Length Analyzer

# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze each message
for msg in messages:
    length = len(msg)   # Calculate message length
    print(f"\nMessage: '{msg}' | Length: {length}")

    # Flag messages longer than 10 characters
    if length > 10:
        print("Flag: Message longer than 10 characters")
```

```python
# 5. Error Message Detector

# System log entries
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Count occurrences of "ERROR"
error_count = logs.count("ERROR")

# Print total error entries
print("\nTotal ERROR entries:", error_count)
```
