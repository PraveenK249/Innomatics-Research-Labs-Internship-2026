```python
# Function to manage parking lot
def manage_parking(capacity, logs):
    parked = 0
    peak_usage = 0

    # Process each vehicle log
    for entry in logs:
        if entry == "IN":
            parked += 1
        elif entry == "OUT" and parked > 0:
            parked -= 1

        # Track peak usage
        if parked > peak_usage:
            peak_usage = parked

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Over Capacity Alert")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak_usage)


# Example
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
manage_parking(capacity, vehicle_logs)
```

```python
# Function to estimate delivery time
def estimate_delivery_time(distance, traffic, weather):
    # Base time assumption (5 minutes per km)
    eta = distance * 5

    # Traffic impact
    if traffic.lower() == "high":
        eta += 15
    elif traffic.lower() == "medium":
        eta += 8

    # Weather impact
    if weather.lower() == "rainy":
        eta += 10
    elif weather.lower() == "stormy":
        eta += 20

    print("Estimated Delivery Time:", eta, "minutes")


# Example
estimate_delivery_time(8, "High", "Rainy")
```

```python
# Function to analyze seat occupancy
def analyze_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)

    occupancy = (booked_count / total_seats) * 100
    print("Occupancy:", round(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Example (simulate 150 bookings)
total = 200
bookings = [1] * 150
analyze_occupancy(total, bookings)
```

```python
# Function to classify server load
def classify_server_load(cpu_readings):
    total = 0

    # Calculate total CPU
    for value in cpu_readings:
        total += value

    average = total / len(cpu_readings)

    print("Average CPU Load:", round(average), "%")

    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Example
cpu_data = [45, 60, 70, 85, 90]
classify_server_load(cpu_data)
```

```python
# Function to monitor classroom resource usage
def monitor_resources(resource_usage):
    overused = []

    # Threshold: more than 8 hours is considered overuse
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Example
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(usage_data)
```

```python
# Function to manage event registration
def manage_registrations(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Example
manage_registrations(100, 105)
```
