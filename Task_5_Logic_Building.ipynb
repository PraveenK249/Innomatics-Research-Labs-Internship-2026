```python
# Function to analyze post engagement
def analyze_engagement(likes_list):
    total_likes = 0

    # Calculate total likes using loop
    for likes in likes_list:
        total_likes += likes

    print("Total Likes:", total_likes)

    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example
likes_data = [200, 350, 150, 400]
analyze_engagement(likes_data)
```

```python
# Function to check medicine stock
def check_medicine_stock(stock):
    print("Medicine Stock:", stock)

    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")


# Example
check_medicine_stock(6)
```

```python
# Function to check rainfall adequacy
def check_rainfall(rainfall_data, required_level):
    total = 0

    # Calculate total rainfall
    for value in rainfall_data:
        total += value

    average = total / len(rainfall_data)
    print("Average Rainfall:", average)

    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example
rainfall_list = [70, 75, 68, 80, 67]
check_rainfall(rainfall_list, 70)
```

```python
# Function to detect duplicate usernames
def detect_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example
users = ["user1", "user2", "user3", "user1"]
detect_duplicates(users)
```

```python
# Function to filter premium crops
def filter_premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)


# Example
crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

```python
# Function to evaluate application health
def check_system_health(error_count):
    print("Error Count:", error_count)

    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# Example
check_system_health(7)
```

```python
# Function to validate transaction amount
def check_transaction(amount):
    daily_limit = 50000
    print("Transaction Amount:", amount)

    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


# Example
check_transaction(60000)
```

```python
# Function to calculate attendance and check eligibility
def check_exam_eligibility(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example
attendance_list = ["P", "P", "A", "P", "P"]
check_exam_eligibility(attendance_list)
```

```python
# Function to calculate electricity bill
def calculate_electricity_bill(units):
    bill = 0

    # Billing slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# Example
total_bill, usage_status = calculate_electricity_bill(250)
print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
