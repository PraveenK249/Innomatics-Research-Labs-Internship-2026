```python
# Function to validate recharge
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        # Check minimum amount
        if amount < 50:
            print("Recharge amount must be at least ₹50. Try again.")
            continue

        # Check if plan is valid
        if amount in valid_plans:
            print("Recharge Successful!")
            break
        else:
            print("Invalid plan. Available plans:", valid_plans)
            print("Please try again.")

#Calling the Function
print(validate_recharge())
```

```python
# Function to check inventory
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")


# Sample data
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 8,
    "Monitor": 20
}

check_inventory(inventory)

```

```python
# Function to process result
def student_result(marks):
    total = 0

    # Calculate total using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Example
marks_list = [60, 45, 70, 55]
print("Result:", student_result(marks_list))
```

```python
# Function to calculate fare
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)

    # Add 25% extra during peak hours
    if peak.lower() == "yes":
        fare += fare * 0.25

    return fare


# Retry system
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

```python
# Function to check attendance eligibility
def check_attendance(attendance):
    present_days = 0

    # Count present days
    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example
attendance_list = ["P", "P", "A", "P", "P"]
print("Status:", check_attendance(attendance_list))
```

```python
# Function to check password strength
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Minimum length check
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    # Loop through characters
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special character)"


# Example
pwd = input("Enter password: ")
print(check_password(pwd))
```
