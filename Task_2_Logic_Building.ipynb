```python
# Problem 1: Unique Words in a Sentence

sentence = "python is easy and python is powerful"

# Split sentence into words
words = sentence.split()

# Convert list to set to remove duplicates
unique_words = set(words)

# Print results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Problem 2: Highest Salary from Employee Data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)

print("\nHighest Salary:", highest_employee, "-", employees[highest_employee])
```

```python
# Problem 3: Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]

# Initialize max and min using first element
max_val = numbers[0]
min_val = numbers[0]

# Compare each value
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("\nList:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

```python
# Problem 4: Count Products Above Price Threshold

prices = [450, 1200, 899, 1500, 300]

count = 0

# Count prices greater than 1000
for price in prices:
    if price > 1000:
        count += 1

print("\nProducts above 1000:", count)
```

```python
# Problem 5: Calculate Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]

# Count number of present days
present_days = attendance.count("P")

# Calculate percentage
percentage = (present_days / len(attendance)) * 100

print("\nAttendance Percentage:", percentage)
```

```python
# Problem 6: Remove Duplicate Phone Numbers

phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert to set to remove duplicates
unique_numbers = set(phone_numbers)

print("\nUnique phone numbers:", unique_numbers)
```

```python
# Problem 7: Count Character Frequency

text = "python"

frequency = {}

# Count frequency of each character
for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print("\nCharacter frequency:", frequency)
```

```python
# Problem 8: Convert List to Tuple

lst = [10, 20, 30]

# Convert list to tuple
tup = tuple(lst)

print("\nTuple:", tup)
```

```python
# Problem 9: Check if Key Exists in Dictionary

emp = {"Ravi": 1, "Anita": 2}

# Check if key exists
if "Ravi" in emp:
    print("\nEmployee exists")
else:
    print("\nEmployee does not exist")
```

```python
# Problem 10: Calculate Average Marks

marks = [70, 75, 65, 80]

# Calculate average
average = sum(marks) / len(marks)

print("\nAverage Marks:", average)
```
